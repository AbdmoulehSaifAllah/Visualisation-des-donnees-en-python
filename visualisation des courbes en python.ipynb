```python
#Importer les bibliothéques voulus
import numpy as np
from scipy.stats import uniform, expon, cauchy
import matplotlib.pyplot as plt
import math

#2)
#Initialisation des variables
n = 1000
lamda = 1

#Générer une distribution uniforme
U = uniform.rvs(size=n)

#Simuler la loi exponentielle
X = []
for i in range (len(U)):
    x = -1*math.log(U[i])/lamda
    X.append(x) # Ajouter la valeur x à la fin de la liste X
    
#3)
#Tracage d'un histogramme des nombres aléatoires simulés
plt.hist(X, density=True)

#Tracage de la fonction de densité de probabilité de la loi de exponentielle
abscisse1 = np.linspace(0,15,n)
ordonnee1 = expon.pdf(abscisse1 ,lamda)
plt.plot(abscisse1, ordonnee1)
plt.show()

#5)
#Simuler la loi de Cauchy
Y=[]
for i in range (len(U)):
    y = lamda*math.tan(math.pi*(U[i]-0.5))
    Y.append(y) # Ajouter la valeur y à la fin de la liste Y

#6)
#Tracage d'un histogramme des nombres aléatoires simulés
plt.hist(Y, density=True)

#Tracage de la fonction de densité de probabilité de la distribution de Cauchy
abscisse2 = np.linspace(-10,10,n)
ordonnee2 = cauchy.pdf(abscisse2 ,lamda)
plt.plot(abscisse2, ordonnee2)
plt.show()



```

```python
#Importer les bibliothéques voulus
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

#Initialisation des variables
nbAccident = np.arange(7)
frequences = [56, 22, 9, 1, 0, 1, 0]

#1
diagBatFreq = [sum(nbAccident == i) for i in range(7)] # diagramme en bâtons de fréquences
plt.bar(nbAccident, frequences, width = 0.5, color = 'lightblue')
plt.show()

#2
sommeFreq = sum(frequences)
diagBatProba = [frequences[i]/sommeFreq for i in range(7)] #diagramme en bâtons de probabilités
plt.bar(nbAccident, diagBatProba, width = 0.5, color = 'orange')
plt.show()

#3
""" Il fait sens de modéliser ces données par un échantillon de loi Poisson d’un certain paramètre λ
car les diagrammes de batons de ces données sont presque les mêmes que le diagramme de baton de la loi de poisson.
En plus, il s'agit de nombre d’accidents d’assurés pour une assurance pendant une période de temps
(dans cet exemple, c'est pendant une année). """

#4
#Remplir la liste
liste = []
for i in range(56):
    liste.append(0)
for i in range(22):
    liste.append(1)
for i in range(9):
    liste.append(2) 
liste.append(3)
liste.append(5)

# le nombre moyen d’accidents
λ_hat = np.mean(liste)

#5
#diagramme en bâtons de probabilités
plt.bar(range(7), diagBatProba, width=0.3,color = 'green')

#La fonction de masse de la loi de Poisson de paramètre λ_hat
vecteur = range(0,7)
masse=poisson.pmf(vecteur ,λ_hat)
vect=np.arange(7)+0.15 #ajout de 0.15 pour bien montrer la visualisation 
plt.bar(vect, masse, width=0.3,color = 'lightgreen')
















```
